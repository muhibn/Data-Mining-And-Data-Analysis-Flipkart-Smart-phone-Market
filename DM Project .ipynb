```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label = le.fit_transform(df['brand'])
df["brand"] = label

label = le.fit_transform(df['base_color'])
df["base_color"] = label

label = le.fit_transform(df['processor'])
df["processor"] = label

label = le.fit_transform(df['screen_size'])
df["screen_size"] = label
df
```

```python
dfcf=df.drop(['model'], axis = 1)
df.columns
```

```python
X = df[[ 'base_color', 'processor', 'screen_size', 'ROM',
       'RAM', 'display_size', 'num_rear_camera', 'num_front_camera',
       'battery_capacity', 'ratings', 'num_of_ratings', 'sales_price',
       'discount_percent', 'sales']]
Y = df[['brand']]

scaler = StandardScaler()
dfstad = scaler.fit_transform(X)
```

```python
df
```

```python
def holdout_method(neighbor_number):
  X_train, X_test, Y_train, Y_test = train_test_split(dfstad, Y, test_size=0.33)
  clf = DecisionTreeClassifier()
  clf = clf.fit(X_train,Y_train)
  Y_pred_clf = clf.predict(X_test)
  Accuracy_clf = accuracy_score(Y_test, Y_pred_clf).round(3)

  knn = KNeighborsClassifier(n_neighbors=neighbor_number)
  knn = knn.fit(X_train,Y_train)
  Y_pred_knn = knn.predict(X_test)
  Accuracy_knn = accuracy_score(Y_test, Y_pred_knn).round(3)

  nb = GaussianNB()
  nb = nb.fit(X_train,Y_train)
  Y_pred_nb = nb.predict(X_test)
  Accuracy_nb = accuracy_score(Y_test, Y_pred_nb).round(3)

  Accuracy_scores = "\nAccuracy of decision tree {} \nAccuracy of k-nearest {} \nAccuracy of Naive Bayes {}"
  print(Accuracy_scores.format(Accuracy_clf, Accuracy_knn, Accuracy_nb))


  def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
  x = ['Decision Tree','K-Nearest','Naive-Bayes']
  y = [Accuracy_clf, Accuracy_knn, Accuracy_nb]
  plt.bar(x, y, color='purple')
  addlabels(x,y)
  plt.xlabel('Classifiers Name')
  plt.ylabel('Accuracy of the models')
  plt.title('Bar Graph for Holdout Method (Classifiers vs Accuracy)')
  plt.show()

def random_subsampling(neighbor_number):
  Final_accuracy_clf = 0
  Final_accuracy_knn = 0
  Final_accuracy_nb = 0
  random_subsampling_value = int(input("\nEnter number of times random subsampling to be perfomed: "))
  for iteration_count in range(1, random_subsampling_value + 1):
    X_train, X_test, Y_train, Y_test = train_test_split(dfstad, Y, test_size=0.33)
    # Applying Decision Tree
    clf = DecisionTreeClassifier()
    clf = clf.fit(X_train,Y_train)
    Y_pred_clf = clf.predict(X_test)
    Accuracy_clf = accuracy_score(Y_test, Y_pred_clf).round(3)
    Final_accuracy_clf += Accuracy_clf
    # Applying K-Nearest
    knn = KNeighborsClassifier(n_neighbors=neighbor_number)
    knn = knn.fit(X_train,Y_train)
    Y_pred_knn = knn.predict(X_test)
    Accuracy_knn = accuracy_score(Y_test, Y_pred_knn).round(3)
    Final_accuracy_knn += Accuracy_knn
    # Applying Naive-Bayes
    nb = GaussianNB()
    nb = nb.fit(X_train,Y_train)
    Y_pred_nb = nb.predict(X_test)
    Accuracy_nb = accuracy_score(Y_test, Y_pred_nb).round(3)
    Final_accuracy_nb += Accuracy_nb

    Accuracy_scores = "\nDecision Tree Accuracy at {} is {} \nk-nearest Accuracty at {} is {} \nNaive Bayes Accuracy at {} is {}"
    print(Accuracy_scores.format(iteration_count,Accuracy_clf, iteration_count,Accuracy_knn, iteration_count,Accuracy_nb))

  final_clf = (Final_accuracy_clf/random_subsampling_value).round(3)
  final_knn = (Final_accuracy_knn/random_subsampling_value).round(3)
  final_nb = (Final_accuracy_nb/random_subsampling_value).round(3)
  print("\nAverage Decison Tree Accuracy: {}\n Average K-nearest Accuracy: {}\n Average Naive Bayes Accuracy: {}".format(final_clf, final_knn, final_nb))

  # Visualization
  def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
  x = ['Decision Tree','K-Nearest','Naive-Bayes']
  y = [final_clf, final_knn, final_nb]
  plt.bar(x, y, color='purple')
  addlabels(x,y)
  plt.xlabel('Classifiers Name')
  plt.ylabel('Accuracy of the models')
  plt.title('Bar Graph for Random-Subsampling (Classifiers vs Accuracy)')
  plt.show()

def cross_validation(neighbor_number):
  Final_accuracy_clf = 0
  Final_accuracy_knn = 0
  Final_accuracy_nb = 0
  count = 1
  n_folds = int(input("Enter number of folds: "))
  kf =KFold(n_splits=n_folds, shuffle=True, random_state=None)
  for train_index, test_index in kf.split(dfstad):
    x_train, x_test = dfstad[train_index], dfstad[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    clf = DecisionTreeClassifier()
    clf = clf.fit(x_train,y_train)
    y_pred_clf = clf.predict(x_test)
    Accuracy_clf = accuracy_score(y_test, y_pred_clf).round(3)
    Final_accuracy_clf += Accuracy_clf
    # Applying K-Nearest
    knn = KNeighborsClassifier(n_neighbors=neighbor_number)
    knn = knn.fit(x_train,y_train)
    y_pred_knn = knn.predict(x_test)
    Accuracy_knn = accuracy_score(y_test, y_pred_knn).round(3)
    Final_accuracy_knn += Accuracy_knn
    # Applying Naive-Bayes
    nb = GaussianNB()
    nb = nb.fit(x_train,y_train)
    y_pred_nb = nb.predict(x_test)
    Accuracy_nb = accuracy_score(y_test, y_pred_nb).round(3)
    Final_accuracy_nb += Accuracy_nb

    Accuracy_scores = "\nDecision Tree Accuracy at {} is {} \nk-nearest Accuracty at {} is {} \nNaive Bayes Accuracy at {} is {}"
    print(Accuracy_scores.format(count,Accuracy_clf, count,Accuracy_knn, count,Accuracy_nb))
    count += 1

  final_clf = (Final_accuracy_clf/n_folds).round(3)
  final_knn = (Final_accuracy_knn/n_folds).round(3)
  final_nb = (Final_accuracy_nb/n_folds).round(3)
  print("\nAverage Decison Tree Accuracy: {}\nAverage K-nearest Accuracy: {}\nAverage Naive Bayes Accuracy: {}".format(final_clf, final_knn, final_nb))

  # Visualization
  def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
  x = ['Decision Tree','K-Nearest','Naive-Bayes']
  y = [final_clf, final_knn, final_nb]
  plt.bar(x, y, color='red')
  addlabels(x,y)
  plt.xlabel('Classifiers Name')
  plt.ylabel('Accuracy of the models')
  plt.title('Bar Graph for Cross-Validation (Classifiers vs Accuracy)')
  plt.show()


def user_choice():
  character = 'y'
  neighbor_number = int(input("Enter number of neighbors for k-nearest classifier: "))
  while character == 'y':
    print("1. Holdout Method \n2. Random Subsampling \n3. Cross-Validation")
    choice = int(input("\nEnter your choice: "))
    if choice == 1:
      holdout_method(neighbor_number)
    elif choice == 2:
      random_subsampling(neighbor_number)
    elif choice == 3:
      cross_validation(neighbor_number)
    else:
      print("Wrong choice!!")
    character = input("\nDo you want to continue(y/n)?: ")

def main():
  user_choice()

if __name__=="__main__":
  main()
```
